```python
# 考試 list comprehension 輸出結果是 點數>3就得一分就是為1,否則就是 -1
import matplotlib.pyplot as plt
from random import randint
randint(1, 6) # 產生單一次亂數結果 請幫忙產生100次
games = [1 if randint(1, 6)>3 else -1 for i in range(100)]
games[:6] # [5, 5, 2, 1, 3, 2] --> [1, 1, -1, -1, -1, -1]
plt.figure(figsize=(12, 3))
plt.plot(np.cumsum(games))
plt.show()
np.cumsum(games)[-1]#  利用 np.cumsum(games) 就可以知道盈虧
```

```python
# 考試 list comprehension
from random import randint
import numpy as np
import matplotlib.pyplot as plt
randint(1, 6) # 產生單一次亂數結果 請幫忙產生100次
games = [randint(1, 6) for i in range(100)]
gamesc = np.cumsum(games)
plt.figure(figsize=(12, 3))
plt.plot(gamesc)
plt.show()
```
