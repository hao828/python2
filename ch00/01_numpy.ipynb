```python
# list  : element data type 可以各自不同
# numpy : element data type 必須一致
# boolean slicing <------------------------
import numpy as np
student = np.array([10, 20, 30, 40, 50])
scores2 = np.array([44, 55, 66, 77, 88])
# 請問 怎樣取得 20 <=student <=40, # 避開使用迴圈
filter = (student >= 20) & (student <=40)
print(filter)
scores2[filter]
```

```python
import numpy as np
student = np.array([10, 20, 30, 40, 50])
scores2 = np.array([44, 55, 66, 77, 88])
# 請問 怎樣取得 20 <=student <=40
scores2[ (student >= 20) & (student <=40) ] #取代回圈
# {}字 ()動作 [] numpy/list
```

```python
# 請問如何將 以下數據 變化成 pandas.DataFrame
# student scores
#   10   44
#   20   55
#   ..   ..
import pandas as pd
import numpy as np
student = np.array([10, 20, 30, 40, 50])
scores2 = np.array([44, 55, 66, 77, 88])
sample = pd.DataFrame({'student': student, 'scores': scores2})
sample
```

```python
# 請問怎樣找sample 的 0, 1, 2 學生 編號 & 分數
sample[:3]
```

```python
sample.iloc[[1, 3, 4],  :]
```

```python
# 從 sample 中的 scores 欄位判斷 大於等於 60分以上者 的 student編號是多少
# 輸出 DataFrame
#     student .. scores
#
sample[ sample['scores'] >= 60 ]
```
